```python
import requests;
from bs4 import BeautifulSoup;
from tqdm import tqdm;
import pandas as pd;
headers = {
    "User-Agent": "KaggleWikiResearch/1.0 (learning project)"
}
```

```python
title=str(input("Enter Topic: ")).replace(" ","+");
gLink="https://www.google.com/search?q="+title+"+wikipedia";
print(gLink)
res=requests.get(gLink)
soup=BeautifulSoup(res.text,"html.parser")
# soup.find("div",class_).find_all("a",class_="zReHs")
for sp in soup.find_all("div"):
    try:
        link=sp.find("a").get("href")
        print(link)
    except:
        pass
```
